## Plotting raw data

```python
container.query('rho_a > 0 and rho_a < 800')
```

```python
import edf.plotters as plotters
plotters.histograms.plot_histograms(container, ['R', 'rho_a', 'Iab', ])
```

```python
def iterate_extra_dims(container, extra_dims):
    g = container.df.groupby(extra_dims)
    for name, group in g:
        yield name, group
    
for extra_dim, subset in iterate_extra_dims(container, ['timestep', ]):
    plotters.histograms.plot_histograms(subset, ['R', 'rho_a', 'Iab', ])
```

```python
plotters.histograms.plot_histograms_extra_dims(container, ['R', 'rho_a', 'Iab', ], extra_dims=['timestep', ])
```

```python
container.df['rho_a'].describe()
```

```python
import crtomo.configManager as CRConfig
crtcfg = CRConfig.ConfigManager(nr_of_electrodes=96)
g = container.df.groupby('timestep')

for timestep, group in g:
    print(timestep)
    configs = group[['A', 'B', 'M', 'N']].values
    crtcfg.clear_configs()
    crtcfg.add_to_configs(configs)
    R = group['R'].values
    mid = crtcfg.add_measurements(R)
    crtcfg.plot_pseudosection_type2(mid, log10=True)
    rhoa = group['rho_a'].values
    mid = crtcfg.add_measurements(rhoa)
    crtcfg.plot_pseudosection_type2(mid)
```
